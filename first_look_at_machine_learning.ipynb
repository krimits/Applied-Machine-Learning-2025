* Wikipedia [defines as Machine learning](https://en.wikipedia.org/wiki/Machine_learning) "a field of inquiry devoted to understanding and building methods that 'learn', that is, methods that leverage data to improve performance on some set of tasks".

* We can distinguish three kinds of Machine Learning:

    * supervised learning
    
    * unsupervised learning
    
    * reinforcement learning

* In supervised learning, we give the computer a set of problems *with their solutions*.

* In the *training* phase, the computer will manage (we hope) to find the way that from a problem we get to its solution.

* The supervision rests on the fact that the computer has a guide: the solutions.

* Once training is over, we want to check how well the computer has learned from the data that we have given it.

* We do that in the *testing* phase.

* Typical applications of supervised learning are:

  * Classification.
  
  * Regression.

* In unsupervised learning, we give the computer only a set of problems, without solutions.

* Typical applications:

  * Clustering.
  
  * Dimensionality reduction.

* In reinforcement learning, the computer somehow interacts with the environment, getting feedback to its behavior.

* When it acts in the right way, it gets a reward; otherwise, it gets a kind of punishment, or fine.

* Such applications are autonomous systems like driverless cars, robots, etc.

## The Road to Support Vector Machines

* We'll introduce supervised Machine Learning through a method called Support Vector Machines.

* We'll get there gradually, step by step.

* Suppose that we have data with two attributes, or dimensions, and the data belong in two different categories, or classes.

<img src="two_classes.png" />

* We want to find a way so that for any point $(x, y)$ we will be able to predict whether it belongs to class 0 or to class 1.

<img src="two_classes_classify_x.png" />

* We can do that by dividing the plane in two using a line.

* Here are three candidate lines.

<img src="two_classes_three_classifiers.png" />

* The question is, which one should we choose?

* We define as *margin* the least distance of a line from our data.

* Here is the margin of each line.

<img src="two_classes_three_classifiers_margins.png" />

* Then the answer to our problem is: we select the line with the largest possible margin.

* Our classifier is then a *Maximal Margin Classifier*.

* That's the line with the pink margin in our example.

<img src="two_classes_classifier_margin.png" />

* Therefore, the solution to our problem is the line:

  $$ y = 0.17 x + 2.32 $$

* We know that a line is defined by:

  $$ y = ax + b $$

* Learning, in our case, consists in finding the two parameters $a = 0.17$ and $b = 2.32$.

* If we think about it a bit, the line is defined only by the points that are closest to it.

* The rest of the points do not really matter.

* The points that define the line are called *support vectors*.

* They are the points that are tangent on the margin.

* Let's see the support vectors in our example:

<img src="two_classes_classifier_support_vectors.png" />

* Let's take stock of where we are.

* We want a way to find, that is, *to learn*, the line that partitions the two classes in the best possible way.

* That means that we want to fidn the line with the widest possible margin.

* How do we find it? Not by trying out lines: there is an algorithm that given the data that we have detects the best possible line.

* Of course, our data do not necessarily have two attributes (dimensions).

* In most cases they will have many more dimensions.

* The same holds, though, for more dimensions.

* If our data have three attributes, or dimensions, then to partition them we need a *plane*.

* If our data have more than three attributes, or dimensions, then to partition them we need to find the appopriate generalisation of a plane, which is called a *hyperplane*.

* This can also be done efficiently with an algorithm, as when we are trying to find a line.

* Let's add in our dataset an observation of $0$ at position $(0, 2)$.

<img src="two_classes_classifier_extra_point.png" />

* Then the line that we had found before no longer partitions our data completely in two classes.

* The optimum classifier is now the line:

  $$ y = 0.61 x  + 1.34$$

* If we draw the new line in black and the previous one in green, we can see the change that was brought by a single observation.

<img src="two_classes_classifier_extra_point_updated.png" />

* But do we really want something like that?

* Do we want the solution to our problem to be so sensitive to small changes?

* That is not such a good idea.

* We might prefer to have some slack: let some observations violate the margin, even the hyperspace, so that our solution is more robust to outliers.

* But we would like to control the slack.

* Indeed, if we do give the algorithm some slack, it will come up with a new line, which classifies the new point wrongly, but we might accept that:

  $$ y = 0.37 x + 2.13 $$

<img src="two_classes_sv_classifier.png" />

* The new, adapted, algorithm is called *Support Vector Classifier*.

* The good news is that this algorithm can also find our classifier very efficiently.

* What's more, in this algorithm we pass as parameters, not just the data that we want to classify, but also how much slack we want to give to the margin.

* In Machine Learning, the parameters that control the workings of an algorithm are called *hyperparameters*.

* Now let's see a different dataset.

* In contrast with the previous one, the data are not *linearly separable*.

<img src="two_classes_nlseparable.png" />

* That means that we cannot find a straight line that divides them.

* Of course, the same can happen in more dimensions, where we will not be able to find the hyperplane that separates our data.

* We can, however, project our data to more dimensions, where we hope that they will be linearly separable.

* We can add a third dimension $z$ using the function:

  $$ z = e^{-(x^2 + y^2)} $$

* The points near $(0, 0)$ will have a $z$ value close to $1$, while the points near the periphery will have lower values for $z$.

* Indeed, we can verify that this is so in the following 3D figure:

<img src="two_classes_nlseparable_3d.png" />

* Once we have found a classifier hyperplane in three dimensions, we can return to the original two dimensions and see how the data are classified there:

<img src="two_classes_nlseparable_2d.png" />

* This generalisation of Support Vector Classifiers are called *Support Vector Machines*.

* Researchers have found functions, like the exponential one we used, that produce good results in handling non linearly separable data.

* These functions are called *kernel functions*.

* Summming up, with Support Vector Machines we have a mechanism with which the computer can learn, from the examples that we provide, to classify our data in two different classes.

* Now we can see Support Vector Classifiers in action.

* This is a function to calculate the margin:

```python
def find_margin(intercept, slope, samples):
    margin = 1000 # above any reasonable value
    for x, y in samples:
        perpendicular_distance = abs((slope * x - y + intercept) / math.sqrt(slope**2 + 1))
        margin = min(perpendicular_distance, margin)
    return margin
```

* We plot it.

```python
g = ggplot() +\
    geom_point(aes(x=samples[:,0], y=samples[:,1], color='factor(labels)')) +\
    labs(color = "class")

colors = ['red', 'green', 'blue']
for (intercept, slope), color in zip(lines, colors):
    margin = find_margin(intercept, slope, samples)
    ys = intercept + slope * samples[:,0]
    g += geom_abline(aes(intercept=intercept, slope=slope))
    g += geom_ribbon(aes(x=samples[:,0], ymin=ys-margin, ymax=ys+margin),
                     alpha=0.1, fill=color)

g.save('two_classes_three_classifiers_margins.png')
```

* The problem can be cast as an *optimization problem*.

* The mathematical formulation is:

  $$ \begin{aligned}
     & \underset{\mathbf{w}}{\operatorname{minimize}} && \frac{1}{2} \| \mathbf{w}\|^{2} \\
     & \text{subject to} && y_{i}(\mathbf{w}^{\intercal}\mathbf{x}_{i}-b)\geq 1\quad \forall i\in \{1,\dots ,n\}
     \end{aligned}
  $$

* Why?

  $$ \mathbf{w}^{\intercal }\mathbf{x}_{i} - b = 0 $$
  
  is the equation defining a hyperplane.

* Note that:

  $$ \mathbf{w}^{\intercal}\mathbf{x}_{i} - b = 1 $$
  
  is the equation describing the boundary for one class, $y_i = 1$, and
  
  $$ \mathbf{w}^{\intercal}\mathbf{x}_{i} - b = -1 $$
  
  is the equation describing the boundary for the other class, $y_i = -1$.

* Therefore:

  $$ y_i(\mathbf{w}^{\intercal}\mathbf{x}_{i} - b) \ge 1 $$

* While the distance between the two boundaries, i.e., the margin is $\frac{2}{\|\mathbf{w}\|}$, we want to maximize that, and maximizing the square is easier in optimization.

* We saw that Support Vector Classifiers are a generalisation of Maximal Margin Classifiers.

* Instead of trying to find the biggest margin, they allow some observations to lie on the wrong side of the margin, even on the wrong side of the hyperplane.

* When we do that, we move from a Maximal Margin Classifier to a Support Vector Classifier.

* This, also, can be cast as an optimization problem.

* The mathematical formulation is:

   $$ \begin{aligned}
     & \underset{\mathbf{w}}{\operatorname{minimize}} && \frac{1}{2} \| \mathbf{w}\|^{2} + C \sum_{i=1}^{n}\epsilon_i\\
     & \text{subject to} && y_{i}(\mathbf{w}^{\intercal}\mathbf{x}_{i}-b)\geq (1 - \epsilon_i) \quad \forall i\in \{1,\dots ,n\}
     \end{aligned}
  $$

* If $\epsilon_i = 0$, then the $i$th observation lies on the correct side of the hyperplane.

* If $\epsilon_i > 0$ then the $i$th observation lies inside the margin.

* If $\epsilon_i > 1$ then the $i$th observation lies on the wrong side of the plane.

* If $C = 0$, there can be no violations.

* If $C > 0$, then no more than $C$ observations can be on the wrong side of the hyperplane.

* The observations that lie on the margin, or on the wrong side of the margin are support vectors, because they affect the classifier (the rest *do not*).

* Note that in scikit-learn, the mathematical formulation is a bit different.

* The hyperparameter `C` is *inversely proportional* to the slack (and not directly proportional like here).

* In other words, in scikit-learn, a lower `C` encourages a *wider* margin.

* Common kernels are:

* The following helper function (from https://jakevdp.github.io/PythonDataScienceHandbook/05.07-support-vector-machines.html) can plot the SVC decision function, specifically the line and the margins.

```python
def plot_svc_decision_function(model, ax=None, plot_support=True):
    """
    Plot the decision function for a 2D SVC.
    Taken from https://jakevdp.github.io/PythonDataScienceHandbook/05.07-support-vector-machines.html
    """
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()

    # create grid to evaluate model
    x = np.linspace(xlim[0], xlim[1], 30)
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)

    # plot decision boundary and margins
    ax.contour(X, Y, P, colors='k',
               levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])

    # plot support vectors
    if plot_support:
        ax.scatter(model.support_vectors_[:, 0],
                   model.support_vectors_[:, 1],
                   s=300, linewidth=1, facecolors='none');
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
```

```python
plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=(8, 6))
plt.scatter(samples[:, 0], samples[:, 1], s=50,
            c=labels, cmap=ListedColormap(['#F8766D', '#00BFC4']))
plot_svc_decision_function(clf)
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1],
            s=200, lw=1, facecolors='none', edgecolor='black')
plt.savefig('two_classes_nlseparable_2d.png')
```

* Before we proceed, let's recall that up to this point we have seen classifiers for two classes.

* But here we have 10 classes (one for each digit).

* How can we handle that?

* There are two multi-class classification strategies:

  * One-versus-One, OvO
  
  * One-versus-All, OvA or One-versus-Rest, OvR.

* In the One-versus-One strategy, if we have $K$ classes we create $\binom{K}{2} = \frac{K(K-1)}{2}$ classifiers.

* Each one of them classifies each observation between two classes.

* We count the results of the classifications for each observation.

* Each observation will be assigned to the class given by the majority of the classifiers.

* Suppose that we have five classes, C1, C2, C3, C4, C5.

* We will create 10 classifiers.

* We are example results for one sample, which we will assign to C2:

| Classifier | Class |
|------------|-------|
|   C1, C2   |  C2   |
|   C1, C3   |  C1   |
|   C1, C4   |  C1   |
|   C1, C5   |  C5   |
|   C2, C3   |  C2   |
|   C2, C4   |  C2   |
|   C2, C5   |  C2   |
|   C3, C4   |  C4   |
|   C3, C5   |  C5   |
|   C4, C5   |  C4   |

* In the One-versus-All strategy, for $K$ classes we will create $K$ classifiers.

* Each one of them classifies an observation between one class and all the rest.

* At the end, for each observation we select the class that classifies it best.

* In Support Vector Machines the selection is done based on the distance from the hyperplane defined by each specific classifier.

* Suppose we have again fine classes, C1, C2, C3, C4, C5.

* We create five classifiers.

* Each classifier gives a value that corresponds to its belief that the observation belongs to the particular class.

* If we have the following values, the observation will be classified as C3:

| Classifier  | Value  |
|-------------|--------|
|   C1, Rest  |   0.7  |
|   C2, Rest  |   0.6  |
|   C3, Rest  |   0.8  |
|   C4, Rest  |   0.5  |
|   C5, Rest  |   0.4  |

* To classify our digits we will use a Support Vector Machine.

* The default kernel function is the exponential and it accepts a parameter $\gamma$ that multiplies the exponent.

* We will use $\gamma = 0.001$.

* We will also set $C = 100$.  Remember that the `C` hyperparameter controls how strict the margin is.

* The higher the value of `C`, the stricter the classifier becomes.

```python
from sklearn import svm

clf = svm.SVC(gamma=0.001, C=100.)
```

* Once we have constructed our classifier, we must train it.

## Evaluating Machine Learning Outcomes

* Of course in practice we cannot gauge the performance of a classifier in this way.

* For that purpose we use specific metrics.

* The most convenient way to see them is the *classification report*:

```python
from sklearn import metrics

print("Classification report for classifier %s:\n%s\n"
      % (classifier, metrics.classification_report(expected, predicted)))
```

* Let's take the contents of the report one by one.

* The *support* is the number of samples in each class.

* To understand the rest of the metrics, we will need the following:

  * TP, True Positive
  * TN, True Negative
  * FP, False Positive
  * FN, False Negative
  * P, Positive, all positive data in the population
  * N, Negative, all negative data in the population

* The *accuracy* gives the percentage of the correct predictions and is defined as:

  $$ \frac{\textrm{TP} + \textrm{TN}}{\textrm{Sample Size}} $$

* We can get it using the `accuracy_score()` method.

```python
metrics.accuracy_score(expected, predicted)
```

* The accuracy is shown at the lower part of the classification report.

```python
print(metrics.classification_report(expected, predicted))
```

* The *precision* gives us the percentage of the true positives over all our positive predictions.

* It is defined as:

  $$\frac{\textrm{TP}}{\textrm{TP} + \textrm{FP}} = \frac{\textrm{TP}}{\textrm{All Positive Predictions}}$$

* We can get it with the `precision_score()` method.

* If we have many classes, to get the precision for each class we must give the parameter `average=None`.

```python
metrics.precision_score(expected, predicted, average=None)
```

* If we want the average of the precision, we can give the parameter `average='macro'`.

```python
metrics.precision_score(expected, predicted, average='macro')
```

* Alternatively, we can give the parameter `average='weighted'`, so that we take the weighted average of the precisions, with each class precision weighted by its support.

* This may differ from the simple average when our data are imbalanced and there is a substantial difference in the representation of classes in our training data.

```python
metrics.precision_score(expected, predicted, average='weighted')
```

* Precision and the two averages are the first column of the classification report.

```python
print(metrics.classification_report(expected, predicted))
```

* The *recall* metric gives us the percentage of the true positives over all the positives that we should have predived.

* It is defined as:

  $$\frac{\textrm{TP}}{\textrm{TP} + \textrm{FN}} = \frac{\textrm{TP}}{\textrm{P}}$$

* We can get recall using the `recall_score()` method, using the same parameters as with `precision_score()`.

```python
print(metrics.recall_score(expected, predicted, average=None))
print(metrics.recall_score(expected, predicted, average='macro'))
print(metrics.recall_score(expected, predicted, average='weighted'))
```

* Recall is shown in the second column of the classification report.

```python
print(metrics.classification_report(expected, predicted))
```

* Precision and recall are very important with imbalanced data.

* Suppose that our data belong in two classes (positive and negative), and that 99.99% are negative and 0.01% are positive.

* Then a classifier who knows nothing, but always says no, will have an accuracy of 99.99%!

* But it will be completely useless.

* Such a classifier would have $\textrm{TP} = 0$, which means that it would have precision and recall equal to zero.

* At the same time, precision and recall measure different things.

* If we create a very strict classifier that detects only a single positive sample, the precision will be 100% because $\textrm{TP} = 1$ and $\textrm{FP} = 0$, but recall will be abysmal.

* If we create a very lenient classifier that classifies everything as positive, the recall will be 100% because  $\textrm{FN} = 0$, but precision will be abysmal.

* $F_1$ combines precision and recall.

* Mathematically, it is the *harmonic mean* of precision and recall.

* The harmonic mean of $n$ numbers $x_1, x_2, \ldots, x_n$ is defined as:

$$H={\frac {n}{{\frac {1}{x_{1}}}+{\frac {1}{x_{2}}}+\cdots +{\frac {1}{x_{n}}}}}$$

* So, $F_1$ is:

$$F_{1}=2\cdot \frac{1}{\tfrac{1}{\mathrm{precision}}+\tfrac{1}{\mathrm{recall}}} = 2\cdot {\frac{\mathrm{precision} \cdot \mathrm{recall}}{\mathrm{precision} +\mathrm{recall}}}$$

* We use $F_1$ instead of a simple average because it punishes extreme values.

* If either precision or recall are zero, $F_1$ will be zero.

* Another way to evaluate the results is through the *confusion matrix*:

```python
print("Confusion matrix:\n%s" % metrics.confusion_matrix(expected, predicted))
```

## Stochastic Gradient Descent

* A nice feature of scikit-learn is that it offers many different Machine Learning methods.

* A popular method is *Stochastic Gradient Descent*.

* Stochastic Gradient Descent works completely differently from Support Vector Machines.

* Internally, it creates a linear function that, taking in the features of our data, calculates the target (class) we want.

* During training, this function will (probably) make wrong predictions.

* We want to minimize this error, which is called *loss*.

* From calculus, we know that in order to minimize a function we use its derivative.

* The derivative of a function at a point gives us the slope, or gradient, of the function at that point.

* Stochastic Gradient Descent tries to modify the function so that it produces the smallest possible loss, using the slope, or gradient, at its point (member of the training dataset).

* The method is called stochastic because during training we feed it with random batches from our dataset.

* We calculate the loss over the batch and we calculate the corrections to our function using derivatives.

* In a multi-variable function, the partial derivatives of the function over the variables are the *gradient*.

* If we want to get into more detail in what is going on, suppose we want to fit a straight line that passes through the origin predicting $y$ from $x$:

  $$ y = w x$$

* We have a training set with observations:
  $$ x_{1},x_{2},\ldots ,x_{n}$$
  
  
* And the corresponding $y$ values:

 $$ y_{1}, y_{2},\ldots ,y_{n} $$

* We will denote our estimates as:

 $$ \hat{y}_{1}, \hat{y}_{2},\ldots ,\hat{y}_{n} $$

* That is, we have:

  $$ \hat{y}_{1} = wx_1$$
  $$ \hat{y}_{2} = wx_2$$
  $$ \vdots$$
  $$ \hat{y}_{n} = wx_n$$
  
* We want to find the $w$ so that our estimates $\hat{y}_i$ are as close as possible to the observed values, $y_i$.

* We start with a random $w$.

* We define a *loss function* that measures how wrong our estimates are.

* We will use the sum of squares as our loss function.

* Note that this is a function of $w$, the parameter we are trying to find:

  $$Q(w) = \sum _{i=1}^{n}\left({\hat {y}_{i}}-y_{i}\right)^{2}=\sum _{i=1}^{n}\left(wx_{i}-y_{i}\right)^{2}$$
  
  
* This $Q(w)$ is the *objective function* that we are trying to minimize.

* We must minimize the objective function.

* To do that, we can take each $x_i$ and calculate the derivative with respect to $w$ of $Q(w)$:

  $$ \frac{d Q(w)}{dw} = \frac{d(wx_i - y_i)^2}{dw} = 2x_i$$

* If we do that for all $x_i$ and take the average we get:

  $$ \frac{1}{n}\sum_{i=1}^{n}\frac{d(wx_i - y_i)^2}{dw} = \frac{2}{n}\sum_{i=1}^{n}x_i$$
  
* Therefore, we'll adjust $w$ as follows, controlling the adjustments of $w$ by way of a *learning rate* $\eta$:

  $$w \leftarrow w - \eta \left(\frac{2}{n}\sum_{i=1}^{n}x_i \right)$$

* Suppose now that we want to fit a straight line that does not pass through the origin:

  $$ y = w_{1} + w_{2}x$$

* We have a training set with observations:
  $$ x_{1},x_{2},\ldots ,x_{n}$$
  
* And the corresponding estimates:

 $$ \hat{y}_1, \hat{y}_2,\ldots ,\hat{y}_n $$

* We want to estimate the loss using sums of squares.

* The objective function then becomes, for $\mathbf{w} = [w_1, w_2]^\intercal$:

  $$ Q(\mathbf{w}) = \sum _{i=1}^{n}\left({\hat{y}_{i}}-y_{i}\right)^{2}=\sum _{i=1}^{n}\left(w_{1}+w_{2}x_{i}-y_{i}\right)^{2}$$

* For each $x_i$ we need to examine the changes in both $w_1$ and $w_2$. This is done through their *partial derivatives*, in what is called *gradient*.

* If we set:

  $$ Q_i({\mathbf{w}}) = (\hat{y}_i - y_i)^2 = \left(w_{1}+w_{2}x_{i}-y_{i}\right)^{2} $$

* We have:

$$
\nabla Q_i(\mathbf{w}) =
\begin{bmatrix}{\frac {\partial Q_i(\mathbf{w})}{\partial w_{1}}}\\
{\frac {\partial Q_i(\mathbf{w})}{\partial w_{2}}}\end{bmatrix}
=
\begin{bmatrix}{\frac {\partial }{\partial w_{1}}}(w_{1}+w_{2}x_{i}-y_{i})^{2}\\{\frac {\partial }{\partial w_{2}}}(w_{1}+w_{2}x_{i}-y_{i})^{2}\end{bmatrix}$$

* The Batch Gradient Descent method updates $\mathbf{w}$ by iteratively taking the average of the update over the whole dataset, until a minimum is reached:

  $$\mathbf{w} \leftarrow \mathbf{w} -\eta \nabla Q(\mathbf{w}) = \mathbf{w}-\eta \frac{1}{n} \sum _{i=1}^{n}\nabla Q_{i}(\mathbf{w})$$


* Choose an initial vector of parameters $\mathbf{w}$ and learning rate $\eta$.

* Repeat until an approximate minimum is obtained:
  * $Q_{t} \leftarrow \mathbf{0}$
  * For $i=1,2,\ldots,n$, do:
    * $Q_{t} \leftarrow Q_{t} + \nabla Q_{i}(\mathbf{w})$
  * $\mathbf{w} \leftarrow \mathbf{w} - \eta\; Q_{t}/n$

* In Stochastic Gradient Descent, we don't update the parameters after going through the whole dataset.

* Instead of that, we update them after taking each point in turn, randomly.

* Choose an initial vector of parameters $\mathbf{w}$ and learning rate $\eta$.

* Repeat until an approximate minimum is obtained (each iteration over the dataset is called an *epoch*):
  * Randomly shuffle examples in the training set.
  * For $i=1,2,\ldots,n$, do:
    * $\mathbf{w} \leftarrow \mathbf{w} -\eta \nabla Q_{i}(\mathbf{w})$

* The last step of the above algorithm is:

$$
\begin{align*}
\begin{bmatrix}
w_1 \\ w_2\\
\end{bmatrix}
&\leftarrow
\begin{bmatrix}
w_1 \\ w_2\\
\end{bmatrix}
- \eta
\nabla Q_i(\mathbf{w}) \\
&=
\begin{bmatrix}
w_1 \\ w_2\\
\end{bmatrix} -
\eta\begin{bmatrix}{\frac {\partial Q(\mathbf{w})}{\partial w_{1}}}\\
{\frac {\partial Q(\mathbf{w})}{\partial w_{2}}}\end{bmatrix}
\\
&=
\begin{bmatrix}
w_1 \\ w_2\\
\end{bmatrix} -
\eta\begin{bmatrix}{\frac {\partial }{\partial w_{1}}}(w_{1}+w_{2}x_{i}-y_{i})^{2}\\{\frac {\partial }{\partial w_{2}}}(w_{1}+w_{2}x_{i}-y_{i})^{2}\end{bmatrix}
\\
&=
\begin{bmatrix}
w_1 \\ w_2\\
\end{bmatrix} -
\eta\begin{bmatrix}2(w_{1}+w_{2}x_{i}-y_{i})\\2x_{i}(w_{1}+w_{2}x_{i}-y_{i})\end{bmatrix}
\end{align*}
$$

* The advantages of Stochastic Gradient Descent are:
    * Efficiency.
    * Ease of implementation (lots of opportunities for code tuning).

* The disadvantages of Stochastic Gradient Descent include:
    * SGD requires a number of hyperparameters such as the regularization parameter and the number of iterations.
    * SGD is sensitive to feature scaling (about which we'll talk in a bit).

* In general, in (stochastic or batch) gradient descent we want to find a linear scoring function that has for each instance of our data $\mathbf{x_i}$ the form:

  $$f(\mathbf{x}_i) = b + w_1 x_{i1} + w_2 x_{i2} + \cdots + w_p x_{ip}$$
  
  where $\mathbf{x}_i$ is an instance of our data with $p$ features.
  
* To do that we need to find the modelling parameters $b, w_1, w_2, \ldots, w_p$.

* If we set $\mathbf{w} = [ w_1, w_2, \ldots, w_p ]^\intercal $, we write:

  $$f(\mathbf{x_i}) = \mathbf{w} \mathbf{x}_i + b$$

* Then we need to minimize the training error for $n$ observations, given by:

  $$ E(\mathbf{w}, b) = \frac{1}{n}\sum_{i=1}^{n}L(y_i, f(\mathbf{x}_i)) + \alpha R(\mathbf{w}) $$
  
* In the above:

  * $L(\cdot)$ is a loss function that measures model (mis)fit
  * $R(\cdot)$ is a regularization term (aka penalty) that penalizes model complexity
  * $\alpha$ is a non-negative hyperparameter
  
* Note that the notion of loss is not specific to Stochastic Gradient Descent.

* For, example, in Support Vector Machines, $L(\cdot)$ is the [hinge loss](https://en.wikipedia.org/wiki/Hinge_loss).

* A common choice for $R(\cdot)$ is the so-called $\ell_2$ norm:

  $$R(\mathbf{w}) = \frac{1}{2}\sum_{i=1}^n w_i^2$$

* Since we mentioned the hinge loss, let's see what it is.

* For a classifier score $f(x) = \pm 1$ and an intended output $y$, the hinge loss of the prediction $f(\mathbf{x})$ is defined as:

  $$ L(y, f(\mathbf{x}))=\max(0,1-f(\mathbf{x})\cdot y)$$
  
* When $f(\mathbf{x})$ and $y$ have the same sign (meaning $f(\mathbf{x})$ predicts the right class) and $|f(\mathbf{x})|\geq 1$ the hinge loss $L(y, f(\mathbf{x})) = 0$.

* Otherwise, the hinge loss increases linearly with $|f(\mathbf{x})|$.

```python
import numpy as np
from matplotlib import pyplot as plt

plt.figure(figsize=(4, 3))
fx = np.linspace(-2, 3, 100)

y = 1
hinge_loss = np.maximum(np.zeros(fx.shape[0]), 1 - y * fx)
plt.plot(fx, hinge_loss, label='intented = 1')

y = -1
hinge_loss = np.maximum(np.zeros(fx.shape[0]), 1 - y * fx)
plt.plot(fx, hinge_loss, label='intented = -1')

plt.xlabel('$f(x)$')
plt.ylabel('hinge loss')
_ = plt.legend()
```

* The ROC curve uses three things:

  * The True Positive Rate, or TPR (the recall):

  $$ \textrm{TPR} = \frac{\textrm{TP}}{\textrm{P}}$$

  * The False Positive Rate, or FPR:

  $$ \textrm{FPR} = \frac{\textrm{FP}}{\textrm{N}}$$

  * A series of decreasing *thresholds* that we apply on the results of the classifier, because the classifier calculates a probability that something belongs to the class. For each threshold, values above the threshold belong to the class.

```python
fig = plt.figure(figsize=(6, 6))
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([-0.05, 1.0])
plt.ylim([0.0, 1.05])
_ = metrics.RocCurveDisplay.from_estimator(classifier, X_test, y_test, ax=fig.gca())
```

* We then plot the $\textrm{TPR}$ against the $\textrm{FPR}$ for different thresholds: we have the $\textrm{TPR}$ on the $y$ axis and the $\textrm{FPR}$ on the $x$ axis.

* In the lower left corner of the plot, the threshold is high: we are very strict in declaring a positive, so both $\textrm{TP}$ and $\textrm{FP}$ are low.

* In the uppper right corner of the plot, the threshold is low: we are very lenient in declaring a positive, so both $\textrm{TP}$ and $\textrm{FP}$ are high.

* The sweet spot is the upper left corner, if we can get there.

* In a ROC we also draw the diagonal line. This shows a classifier with no discriminatory power: its $\textrm{TPR}$ is equal to its $\textrm{FPR}$.

* For example, if $\textrm{TPR} = \textrm{FPR} = 0.2$, then the classifier is no better than a completely random classifier.

* As we said, the ideal ROC curve hugs the upper left corner.

* The AUC is the area under the curve; the ideal curve has AUC equal to 1. The diagonal has AUC equal to 0.5.

* The diagonal line is the *line of no discrimination* and we said the classifier is not better than a completely random classifier. Here is why (https://datascience.stackexchange.com/a/31877).

* Assume the classifier assigns randomly to the positive class with probability $\rho$ and to the negative with probability $1 - \rho$.

* Also assume that the fraction of positives is $x$ and of negatives $1 - x$.

* Then we have:

  $$ \textrm{TP} =\rho x$$

  $$ \textrm{FP} = \rho (1 - x) $$

  $$ \textrm{FN}  = (1 -\rho) x $$

  $$ \textrm{TN} = (1 - \rho)(1  x) $$

* But then:

  $$ \textrm{TPR} = \frac{\textrm{TP}}{\textrm{TP + FN}} = \frac{\rho x}{(\rho x) + (1 - \rho)x} = \rho $$

  $$ \textrm{FPR} = \frac{\textrm{FP}}{\textrm{TN + FP}} = \frac{\rho (1 - x)}{(1 - \rho)(1 - x) + \rho(1 - x)} = \rho $$

* So for a random classifier, the $\textrm{TPR}$ and $\textrm{FPR}$ are equal, and do not depend on the base rate $x$.

* We can investigate the situation further, by trying different values of $\gamma$.

* We'll try $\gamma = 10^{-1}$, $\gamma = 10^{-2}$, $\gamma = 10^{-3}$.

```python
fig = plt.figure(figsize=(6, 6))
for i in range(3):
    gamma = pow(10, -(i+1))
    classifier = svm.SVC(kernel='rbf',
                         gamma=gamma,
                        random_state=42)
    classifier.fit(X_train, y_train)
    metrics.RocCurveDisplay.from_estimator(classifier, X_test, y_test, ax=fig.gca())
plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([-0.05, 1.0])
plt.ylim([0.0, 1.05])
```

* Finally, let's use a linear kernel.

```python
fig = plt.figure(figsize=(6, 6))
classifier = svm.SVC(kernel='linear',
                     random_state=42)
classifier.fit(X_train, y_train)
metrics.RocCurveDisplay.from_estimator(classifier, X_test, y_test, ax=fig.gca())
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([-0.05, 1.0])
plt.ylim([0.0, 1.05])
```

## Hyperparameter Search

* A crucial part of using Machine Learning is finding the right hyperparameter values.

* But how can we find these values?

* We can ask the computer to help.

* Let's go back to the breast cancer dataset.

```python
bc = datasets.load_breast_cancer(as_frame=True)
bc.data
```

* We would like to use Support Vector Machines, but we have no idea which hyperparameter values to use.

* We create a data structure with the values that we want to check for every hyperparameter we want to control.

* We try different kernels, for `rbf` we try different `gamma` values, and we also want to try one `C` for both classes, or to adjust the `C` values taking into account class frequencies (`class_weight` hyperparameter).

```python
param_grid = [
    {
        "kernel": ["rbf"],
        "gamma": [1e-2, 1e-3, 1e-4],
        "C": [1, 10, 100, 1000],
        "class_weight": [None, "balanced"]
    },
    {
        "kernel": ["linear"],
        "C": [1, 10, 100, 1000],
        "class_weight": [None, "balanced"]
    },
]
```

* Then we use the `GridSearchCV` class, which will try out all possible combinations of hyperparameter values and identify the best model.

* As the name`GridSearchCV` implies, it performs $k$-fold validation, where $k = 5$ unless we say otherwise.

* Moreover, and this applies to other scikit-learn methods that can run in parallel, it takes the parameter `n_jobs` with which we ask for a particular number of parallel jobs (or -1 for as many as possible).

```python
from sklearn.model_selection import GridSearchCV

clf = svm.SVC()
grid_search = GridSearchCV(clf, param_grid=param_grid, n_jobs=-1)
grid_search.fit(bc.data, bc.target)
```

* We can see the best values of the hyperparameters from the `best_params_` attribute.

```python
grid_search.best_params_
```

* If we want to use the best model, which was built with these hyperparamer values, we can get it from the `best_estimator_` attribute.

```python
grid_search.best_estimator_
```

* If we want to see all the results of the search, we can get them from the attribute `cv_results_`.

```python
grid_search_results = pd.DataFrame(grid_search.cv_results_)
print(grid_search_results.shape)
grid_search_results
```

* Perhaps it would be better to put them in decreasing order of their score.

```python
grid_search_results.sort_values(by='rank_test_score')
```

* If we just want the best score:

```python
grid_search.best_score_
```

* We can also try an`SGDClassifier`.

* We'll use another approach for hyperparameter searching.

* Instead of performing an exhaustive search, we'll sample randomly among the combinations of hyperparameter values.

* The `average` hyperparameter determines whether we'll use the weights found at the end of SGD, or whether we'll use the average of the weights taken over all updates.

* The `l1_ratio` hyperparameter is the ratio of L1 and L2 regularization that will be used, when for penalty we use `elasticnet`:
  $$ R(\mathbf{w}) := \frac{\rho}{2} \sum_{j=1}^{n} w_j^2 + (1-\rho) \sum_{j=1}^{m} |w_j| $$

* The `alpha` hyperparameter is the constant by which we multiply the regularization term.

```python
import scipy.stats as stats

param_dist = {
    "average": [True, False],
    "l1_ratio": stats.uniform(0, 1),
    "alpha": stats.loguniform(1e-2, 1e0),
}
```

* To do that, we use the `RandomizedSearchCV` class, giving the number of hyperparameter settings to be tried with the `n_iter` parameter.

```python
from sklearn.model_selection import RandomizedSearchCV

clf = SGDClassifier(penalty='elasticnet', max_iter=10**5)
random_search = RandomizedSearchCV(clf,
                                   param_distributions=param_dist,
                                   n_iter=15,
                                   n_jobs=-1)
random_search.fit(bc.data, bc.target)
```

* Let's see what we got:

## Hyperparameter Search with Optuna

* [Optuna](https://optuna.org/) is a hyperparameter optimization framework that can offer additional capabilities to those provided by scikit-learn.

* This time we'll use try three different scalers as well:
    * `MinMaxScaler`
    * `StandardScaler`
    * `RobustScaler`

* `RobustScaler` is less sensitive to outliers than `StandardScaler`. The transformation it performs is:

$$ x_{\text{scaled}} = \frac{x - \text{Median}(X)}{\text{IQR}(X)} $$

```python
import optuna

from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler

def objective(trial):

    scaler_name = trial.suggest_categorical("scaler",
                                            ["minmax", "standard", "robust"])

    if scaler_name == "minmax":
        scaler = MinMaxScaler()
    elif scaler_name == "standard":
        scaler = StandardScaler()
    elif scaler_name == "robust":
        scaler = RobustScaler()

    sgd_average = trial.suggest_categorical("average", [True, False])
    sgd_l1_ratio = trial.suggest_float("l1_ratio", low=0, high=1, step=0.1)
    sgd_alpha = trial.suggest_float("alpha", low=1e-2, high=1, log=True)

    clf = SGDClassifier(penalty='elasticnet', max_iter=int(1e6))

    pipeline = make_pipeline(scaler, clf)

    score = cross_val_score(pipeline, bc.data, bc.target, scoring='accuracy')
    return score.mean()
```

```python
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=100)
```
